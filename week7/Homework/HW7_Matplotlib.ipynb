### P1. Write a Python code to output the following graph. (or a similar one)
```
The productivity values for the 3 groups are the following: 

g1 = [80, 60, 80, 67, 34]
g2 = [45, 67, 23, 89, 10]
g3 = [30, 90, 50, 20, 90]
```

## Problem 1 

````python
import matplotlib.pyplot as plt
import numpy as np

g1 = np.array([80, 60, 80, 67, 34])
g2 = np.array([45, 67, 23, 89, 10])
g3 = np.array([30, 90, 50, 20, 90])

bar_width = 0.18
x_axis = np.array(['M','T','W','Th','F'])
x_axis2 = np.array([0,1,2,3,4])+bar_width
x_axis3 = np.array([0,1,2,3,4])+bar_width*2

plt.title('Productivity during the week')
plt.xlabel('Groups')
plt.ylabel('Productivity')
plt.yticks(np.arange(0,100,20))
plt.ylim(0,100)

plt.bar(x_axis,g1,width=bar_width,label="Group1")
plt.bar(x_axis2,g2,width=bar_width,label="Group2")
plt.bar(x_axis3,g3,width=bar_width,label="Group3")
plt.legend()

plt.show()
````

````python
x = np.random.randint(0,30,25)
y = np.random.randint(0,30,25)

np_colors = np.array(['b', 'g', 'r', 'c', 'm', 'y', 'k'])
colors = np.random.choice(np_colors,25)
sizes = np.random.randint(20,500,25)

plt.title('Colorful dots')
plt.xlabel('X')
plt.ylabel('Y')

plt.xlim(0,32)
plt.ylim(0,32)

plt.scatter(x,y,s=sizes,c=colors)
````
